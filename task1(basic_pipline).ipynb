```python
from transformers import pipeline
```

```python
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")
```

```python
candidate_labels = ['education', 'politics' , 'business']
text = "This is a course about Transformers library"
```

```python
res = classifier(text, candidate_labels)
```

```python
res
```

```python
res['labels']
```

```python
res['scores']
```

```python
res_pobability = res['scores']
max_index = res_pobability.index(max(res_pobability))
max_index
```

```python
res['labels'][max_index]
```

```python
#!pip install transformers[sentencepiece]
from transformers import pipeline
hypothesis_template = "This text is about {}"
classes_verbalized = candidate_labels
zeroshot_classifier = pipeline("zero-shot-classification", model="MoritzLaurer/deberta-v3-large-zeroshot-v2.0")  # change the model identifier here
output = zeroshot_classifier(text, classes_verbalized, hypothesis_template=hypothesis_template, multi_label=False)
print(output)
```

```python
def get_highest_label(output):
    """
    This function takes an output dictionary with 'labels' and 'scores'
    and returns the label with the highest score.
    """
    max_index = output['scores'].index(max(output['scores']))
    return output['labels'][max_index]
```

```python
get_highest_label(output)
```
